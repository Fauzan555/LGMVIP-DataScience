# Importing necessary libraries to perform the task

```python
import cv2
from matplotlib import pyplot as plt
```

# Now reading the image

```python
image = cv2.imread("C:/Users/DEll/Downloads/Fauzan.jpg.jpg")
```

# Now converting the image into gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

# Now inverting the gray scale image

```python
inverted_image = 255 - gray_image
```

# Now making the image blure using Gaussian Blur function 

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

# Now inverting the already blurred image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

# Now plotting the original image

```python
plt.figure(figsize=(15,10))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

# Now making the Pencil sketch of image

```python
plt.figure(figsize=(15,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.show()
```

# Finally we have made pencil sketch of image
